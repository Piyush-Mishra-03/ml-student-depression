# Machine Learning

```python
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier

from sklearn.decomposition import PCA
print('Done')
```

```python
depression.rename(columns={
    'Have you ever had suicidal thoughts ?':'Suicidal_thoughts',
    'Family History of Mental Illness': 'Mental Illness'
    },inplace=True)
```

```python
le = LabelEncoder()
depression['Mental Illness'] = le.fit_transform(depression['Mental Illness'])
depression['FiNaNcial Stress'] = le.fit_transform(depression['FiNaNcial Stress'])
depression['Suicidal_thoughts'] = le.fit_transform(depression['Suicidal_thoughts'])
depression['Degree'] = le.fit_transform(depression['Degree'])
depression['Dietary Habits'] = le.fit_transform(depression['Dietary Habits'])
depression['Profession'] = le.fit_transform(depression['Profession'])
depression['City'] = le.fit_transform(depression['City'])
depression['Gender'] = le.fit_transform(depression['Gender'])

depression.head(150)
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(depression)
print(x_scaled)
```

```python
x = depression.drop('Depression', axis=1)


y = depression['Depression']
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
print(x_pca)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y,test_size=0.3)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train
```

```python
y_train
```

```python

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python

print('Logistic Regression Accuracy:',accuracy_score(y_test,y_pred)*100)
```

```python
rf = RandomForestClassifier(n_estimators=50)
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)

print('Random Forest Accuracy:',accuracy_score(y_test,y_pred_rf)*100)
```

# üìù Model Performance Summary

The classification report below provides precision, recall, and F1-score for both classes.


```python
print(f"‚úÖ Accuracy: {accuracy_score(y_test, y_pred):.2%}")
print("\nüìã Classification Report:")
print(classification_report(y_test, y_pred, labels=[0, 1, 2]))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig('Confusion.png')
plt.show()
```

```python
# Convert target to numeric
y_numeric = pd.factorize(y)[0]

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y_numeric, cmap='coolwarm', edgecolor='k', s=80)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: 2D Projection of Heart Failure Data')
plt.colorbar(label='Target Classes')
plt.tight_layout()
plt.savefig('Confusion.png')
plt.show()
```

## üìä Comparing Model Accuracy
